```python
wiki = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
```

<p> While this might seem like a job for beautifulsoup, pandas has a build in function for reading html </p>

```python
toronto_df = pd.read_html(wiki, header = 0)
```

<p> The previous command produced a list of dataframes, but I only the the first dataframe. </p>

```python
toronto_df[0].head(10)
```

```python
toronto_df = toronto_df[0]
```

```python
toronto_df.head()
```

<p> I need to get rid of the rows containing "Not assigned". In order to do that I subset by data according to the following logical conditions. </p>

```python
toronto_df = toronto_df[(toronto_df.Borough != "Not assigned") & (toronto_df.Neighbourhood != "Not assigned")]
```

```python
toronto_df.head(10)
```

<p> The following command resets the row indexes </p>

```python
toronto_df = Toronto_df.reset_index(drop=True)
```
